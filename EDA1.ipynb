```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv('Cardiotocographic.csv')
data
```

```python
# 1. Handling Missing Values
# Checking missing values
missing_values = data.isnull().sum()
missing_values
```

```python
# Impute missing values with the mean
data_filled = data.fillna(data.mean())
data_filled
```

```python
# 2. Removing Invalid Class Values
data_cleaned = data_filled[data_filled['NSP'].isin([1, 2, 3])]
data_cleaned
```

```python
# 3. Removing Outliers using Z-score
from scipy.stats import zscore
z_scores = np.abs(zscore(data_cleaned))
z_scores
```

```python
data_no_outliers = data_cleaned[(z_scores < 3).all(axis=1)]
data_no_outliers
```

```python
# 4. Statistical Summary
summary_statistics = data_no_outliers.describe()
summary_statistics
```

```python
# 5. Visualizations
plt.figure(figsize=(10, 6))
sns.heatmap(data_no_outliers.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Histogram of the target variable NSP
plt.figure(figsize=(6, 4))
sns.countplot(x='NSP', data=data_no_outliers)
plt.title("Class Distribution")
plt.show()
```

```python
# Box plot to check for outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=data_no_outliers, orient="h")
plt.title("Box Plot for Outlier Detection")
plt.show()
```

```python
# 6. Checking for class imbalance
class_distribution = data_no_outliers['NSP'].value_counts()
class_distribution
```
